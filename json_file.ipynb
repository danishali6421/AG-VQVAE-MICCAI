```python
import json
import os

output_path = "train.json"
modalities = ["t1c", "t1n", "t2f", "t2w"]
data_path = "./data_split_MICAI/train"

data_entries = []

for folder in os.listdir(data_path):
    new_dirname = (os.path.dirname(data_path)).replace("data_split_MICAI", "data_split_MICAI_2025")
    print(new_dirname)
    folder_path = os.path.join(data_path, folder)
    if os.path.isdir(folder_path):
        image_paths = [
            os.path.join(new_dirname, f"{folder}/{folder}-{modality}.nii.gz")
            for modality in modalities
        ]
        print(image_paths)
        mask_path = os.path.join(new_dirname, f"{folder}/{folder}-seg.nii.gz")
        entry = {
            "image": image_paths,
            "mask": mask_path
        }
        data_entries.append(entry)

# Save to JSON
with open(output_path, 'w') as f:
    json.dump(data_entries, f, indent=2)

print(f"Saved {len(data_entries)} entries to {output_path}")
```

```python
import json

# Paths to the individual JSON files
train_json = "train.json"
val_json = "val.json"
test_json = "test.json"

# Load all three
with open(train_json, 'r') as f:
    train_data = json.load(f)

with open(val_json, 'r') as f:
    val_data = json.load(f)

with open(test_json, 'r') as f:
    test_data = json.load(f)

# Merge into one dictionary
combined_data = {
    "train": train_data,
    "val": val_data,
    "test": test_data
}

# Save to one master JSON file
output_path = "all_data_split.json"
with open(output_path, 'w') as f:
    json.dump(combined_data, f, indent=2)

print(f"Merged JSON saved to {output_path}")
```
